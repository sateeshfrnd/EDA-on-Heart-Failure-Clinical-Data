# Exploratory Data Analysis (EDA) on  Heart Failure Clinical Data

## Introduction

The dataset provided for this project is focused on heart failure, a critical medical condition characterized by
the heart's inability to pump blood effectively, leading to inadequate circulation throughout the body. 

### Dataset

The dataset was collected at the Faisalabad Institute of Cardiology and the Allied Hospital in Faisalabad, Punjab, Pakistan, during the period of April to December in 2015. 

The dataset comprises medical records of 299 patients diagnosed with heart failure. These records include various clinical, body, and lifestyle information of the patients. The dataset encompasses 13 features, providing insights into the patients' health status and potential risk factors for heart failure.

### Features

- Age: The age of the patient (numeric).
- Anaemia: Indicates the presence of anemia (binary: 0 for no, 1 for yes).
- Creatinine Phosphokinase (CPK): Level of creatinine phosphokinase enzyme in the blood (numeric).
- Diabetes: Indicates whether the patient has diabetes (binary: 0 for no, 1 for yes).
- Ejection Fraction: Percentage of blood leaving the heart at each contraction (numeric).
- High Blood Pressure: Indicates whether the patient has high blood pressure (binary: 0 for no, 1 for yes).
- Platelets: Platelet count in the blood (numeric).
- Serum Creatinine: Level of creatinine in the blood (numeric).
- Serum Sodium: Level of sodium in the blood (numeric).
- Sex: Gender of the patient (binary: 0 for female, 1 for male).
- Smoking: Indicates whether the patient smokes (binary: 0 for no, 1 for yes).
- Time: Follow-up period (in days) for the patient's condition (numeric).

### Data Cleaning

```python
df = raw_data
```

```python
for col in df.columns:
    print(f"{col} = {df[col].unique()}")
```

```python
df['anaemia'] = df['anaemia'].map({0 : 'No' , 1 : 'Yes'})
df['diabetes'] = df['diabetes'].map({0 : 'No' , 1 : 'Yes'})
df['high_blood_pressure'] = df['high_blood_pressure'].map({0 : 'No' , 1 : 'Yes'})
df['sex'] = df['sex'].map({0 : 'Female' , 1 : 'Male'})
df['smoking'] = df['smoking'].map({0 : 'No' , 1 : 'Yes'})
df['death_event'] = df['death_event'].map({0 : 'No' , 1 : 'Yes'})
```

- Sex: Gender of the patient (binary: 0 for female, 1 for male).
- Smoking: Indicates whether the patient smokes (binary: 0 for no, 1 for yes).

### Data Analysis

#### 1. What is the distribution of age among heart failure patients in the dataset ?

```python
plt.figure(figsize=(8, 6))  # Set the figure size
# plt.hist(df['age'], bins=10, color='skyblue', edgecolor='black', alpha=0.7)
sns.histplot(df['age'], kde = True, color='skyblue', edgecolor='black', alpha=0.7)

# Highlight the mean with a vertical line
mean_value = np.mean(df['age'])
plt.axvline(mean_value, color='red', linestyle='dashed', linewidth=1)

# Highlight standard deviation with vertical lines
std_value = np.std(df['age'])
plt.axvline(mean_value - std_value, color='green', linestyle='dashed', linewidth=1, label='Mean - Std')
plt.axvline(mean_value + std_value, color='green', linestyle='dashed', linewidth=1, label='Mean + Std')


# Add labels and title
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.title('Distribution of Ages')
# plt.grid(True)
plt.show()
```

*Observations:*
- Mean of **age** column is around **60-61**. 
- Most of the people are around **49-72** age.
- Low number of people are above **70** age.

#### 2. How does the death rate vary with age?

```python
# Group ages into decades and set to a new column 'Decode'
df['decade'] = (df['age'] // 10) * 10  
df[['age','decade']]
```

```python
# Group the data by decade and calculate the proportion of deaths for each decade group
decade_death_proportion = df.groupby('decade')['death_event'].value_counts(normalize=True).unstack().fillna(0)
decade_death_proportion
```

```python
# Line plot to visualize how the proportion of deaths varies with age.
plt.figure(figsize=(10, 6))
decade_death_proportion['Yes'].plot(marker='o', linestyle='-', color='red', label='Death Rate')
plt.title('Death Rate by Decade')
plt.xlabel('Decade')
plt.ylabel('Proportion of Deaths')
plt.xticks(decade_death_proportion.index)  # Set x-axis ticks to the decade values
plt.grid(True, linestyle=':', linewidth=1.5, alpha=0.5)
plt.legend()
plt.show()
```

*Observations:*
- Death rate increases with age.

#### 6. How does the prevalence of high blood pressure differ between male and Female patients

```python
# Calculate the total number of patients in each gender group
total_patients = df.groupby('sex').size()
total_patients
```

```python
# Calculate the number of patients with high blood pressure in each gender group
high_bp_patients = df[df['high_blood_pressure'] == 'Yes'].groupby('sex').size()
high_bp_patients
```

```python
# Calculate the proportion of patients with high blood pressure in each gender group
prevalence = (high_bp_patients / total_patients) * 100
prevalence
```

```python
# Create a bar plot
plt.figure(figsize=(6, 4))
sns.barplot(x=prevalence.index, y=prevalence.values, palette='Set2')

# Add labels and title
plt.xlabel('Sex')
plt.ylabel('Prevalence of High Blood Pressure (%)')
plt.title('Prevalence of High Blood Pressure Between Male and Female Patients')

# Show the plot
plt.show()
```

*Observations:*
- We can see that the Female patients have the prevalence of high blood pressure compare with Male.
- *41%* Females and *31%* Males have high blood pressure.

#### 7. What is the relationship between smoking habits and the occurrence on heart failure

```python
# Calculate the total number of patients in each smoking category
total_patients = df.groupby('smoking').size()
total_patients
```

```python
# Calculate the number of patients with heart failure in each smoking category
heart_failure_patients = df[df['death_event'] == 'Yes'].groupby('smoking').size()
heart_failure_patients
```

```python
# Calculate the proportion of patients with heart failure in each smoking category
occurrence = (heart_failure_patients / total_patients) * 100
occurrence
```

```python
# Print the occurrence of heart failure for smokers and non-smokers
print("Occurrence of heart failure having smoking habit :", occurrence['Yes'])
print("Occurrence of heart failure not having smoking habit :", occurrence['No'])
```

```python
# Create a bar plot
plt.figure(figsize=(6, 4))
plt.pie(occurrence, labels=occurrence.index, autopct='%1.1f%%', colors=['lightblue', 'lightgreen', 'lightcoral'])

plt.title('Occurrence of Heart Failure by Smoking Habits')

# Show the plot
plt.show()
```

```python
smoking_proportion = df.smoking.value_counts(normalize=True)*100
smoking_proportion
```

```python
smoking_proportion.plot.pie(
    autopct='%1.1f%%', 
    colors=['lightcoral',  'lightblue'], 
    wedgeprops = dict(width=0.4), 
    startangle = 90)

plt.title('Distribution of Smoking Habits of patients')
plt.show()
```
